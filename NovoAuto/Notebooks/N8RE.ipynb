# N8 AB

```python
AB = pd.read_csv(r'C:\Users\mateo\Documents\NovoAuto\NovoAuto\Input\N8RE2.csv')
AB.columns = AB.columns.map(lambda x: '')
customers_name = AB.iloc[:,2]
```

```python
AB.iloc[125:153,:]
```

### Extraction Customers

```python
customs = []
rangos = []
for i in range(126,157):
    rangos.append(i)
    customs.append(customers_name[i])
customs = [x for x in customs if customs.count(x) == 1] 
customers = {custom: rango for custom, rango in zip(customs, rangos)}
```

```python
customers
```

### Dates

```python
date_spec = []
for a in range(2022,2028):
    for m in range(1,13):
        if m < 10:
            date = '01' + '/' + '0' + str(m) + '/' + str(a)
            date_spec.append(date)
        else:
            date = '01' + '/' +  str(m) + '/' + str(a)
            date_spec.append(date)
```

### Generate of tables

```python
combinaciones = [(cliente, dates) for cliente in customs for dates in date_spec]
ab_output = pd.DataFrame(combinaciones, columns=['customer_name','date_spec'])
```

```python
customs[1]
```

### Target Date

```python
years = {
  '22' : slice(16,28),
  '23' : slice(29,41),
  '24' : slice(42,54),
  '25' : slice(56,68),
  '26' : slice(92,104)
  }  
```

### Add Volume

```python
re_report = []
for custom in customers:
    re_volumenes = []
    re_volume = []
    for year in years:
        re_volumenes.append(AB.iloc[customers[custom],years[year]])
    re_volumenes = pd.concat(re_volumenes, ignore_index=True)
    data = ab_output.loc[ab_output['customer_name'] == custom]
    for valor in re_volumenes:
        re_volume.append(valor)
    data.loc[:, 'volume'] = pd.DataFrame(re_volume)
    re_report.append(data)
re_report = pd.concat(re_report, ignore_index=True)
re_report['use_type'] = 'null'
re_report['brand_name'] = 'N8'
re_report['uom_code'] = 'UI'
re_report['version_type'] = 'RE2'
re_report = re_report[['customer_name', 'date_spec', 'use_type','volume','brand_name','uom_code','version_type']]
```
