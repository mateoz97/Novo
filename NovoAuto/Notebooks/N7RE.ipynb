# N7 RE2

```python
RE2 = pd.read_excel(r'C:\Users\mateo\Documents\NovoAuto\NovoAuto\Input\RE.xlsx', sheet_name='NovoSeven_RE2_Original')
RE2.columns = RE2.columns.map(lambda x: '')
customers_name = RE2.iloc[:,0]
```

### Extraction Customers

```python
customs = []
rangos = []
for i in range(246,589,11):
    rango = slice(i+1,i+10)
    rangos.append(rango)
    customs.append(customers_name[i])
for i in range(588,616):
    customs.append(customers_name[i])
customs = [x for x in customs if customs.count(x) == 1] 
customers = {custom: rango for custom, rango in zip(customs, rangos)}
```

```python
customs
```

### Extraction use_type

```python
use_type= list(RE2.iloc[247:256,0])
```

### Dates

```python
date_spec = []
for a in range(2022,2028):
    for m in range(1,13):
        if m < 10:
            date = '01' + '/' + '0' + str(m) + '/' + str(a)
            date_spec.append(date)
        else:
            date = '01' + '/' +  str(m) + '/' + str(a)
            date_spec.append(date)
```

### Generate of tables

```python
combinaciones = [(cliente, dates, ut) for cliente in customs for dates in date_spec for ut in use_type]
re_output = pd.DataFrame(combinaciones, columns=['customer_name','date_spec','use_type'])
```

### Target Date

```python
years = {
  '22' : slice(40,52),
  '23' : slice(53,65),
  '24' : slice(66,78),
  '25' : slice(79,91),
  '26' : slice(92,104),
  '27' : slice(105,117)  
  }  
```

### Add Volume

```python
re_report = []
for custom in customers:
    re_volumenes = []
    re_volume = []

    for year in years:
        re_volumenes.append(RE2.iloc[customers[custom],years[year]].reset_index(drop=True))
    re_volumenes = pd.concat(re_volumenes, ignore_index=True)
    re_volumenes.columns = columna=['01','02','03','04','05','06','07','08','09','10','11','12']
    data = re_output.loc[re_output['customer_name'] == custom]
    for columna in re_volumenes.columns:
        for valor in re_volumenes[columna]:
            re_volume.append(valor)
    data.loc[:, 'volume'] = pd.DataFrame(re_volume)
    re_report.append(data)
re_report = pd.concat(re_report, ignore_index=True)
re_report['brand_name'] = 'N7'
re_report['uom_code'] = 'MG'
re_report['version_type'] = 'RE2'
```
